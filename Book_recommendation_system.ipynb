```python
rwn = ratings.merge(books, on = 'ISBN')
```

```python
rdf = rwn.groupby('Book-Title').count()['Book-Rating'].reset_index()
rdf.rename(columns = {'Book-Rating':'num_ratings'}, inplace = True)
rdf
```

```python
adf = rwn.groupby('Book-Title').mean()['Book-Rating'].reset_index()
adf.rename(columns = {'Book-Rating':'avg_ratings'}, inplace = True)
adf
```

```python
popular_df = rdf.merge(adf, on = 'Book-Title')
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending = False)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M','num_ratings','avg_ratings']]
```

Collaborative filtering recommendation system,
users who have rated atleast 200 books, 
books which have atleast 50 ratings

```python
X = rwn.groupby('User-ID').count()['Book-Rating'] > 200
study = X[X].index
```

```python
filtered = rwn[rwn['User-ID'].isin(study)]
```

```python
filtered
```

```python
Y = filtered.groupby('Book-Title').count()['Book-Rating'] >=50
famous = Y[Y].index
```

```python
final = filtered[filtered['Book-Title'].isin(famous)]
```

```python
pt = final.pivot_table(index = 'Book-Title',  columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

so now every book is a vector in a 810 dimension space

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
from operator import index
def recommend(book_name):
  index = np.where(pt.index == book_name)[0][0]
  distances = similarity_score[index]
  similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x: x[1], reverse = True)[1:6]
  for i in similar_items:
    print(pt.index[i[0]])
```

```python
recommend('The Da Vinci Code')
```

```python
recommend('1984')
```
